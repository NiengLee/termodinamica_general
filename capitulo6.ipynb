<h1 style="text-align:center;">Capítulo 6 - Segunda ley de la termodinámica</h1>
<p style="text-align:center;">Según la Primera Ley de la Termodinámica, la energía no se crea ni se destruye; únicamente se transforma. En consecuencia, al formular un balance de energía para un sistema, la suma de las entradas y salidas debería conservarse.

<p style="text-align:center;">Sin embargo, al contrastar estos balances con el comportamiento real de los equipos, los resultados ideales no coinciden exactamente con los valores medidos. Por ejemplo, una turbina entrega menos trabajo del previsto teóricamente, o una bomba requiere más energía de la estimada en un análisis ideal.</p>

<p style="text-align:center;">Estas diferencias no contradicen la Primera Ley, sino que se explican mediante la Segunda Ley de la Termodinámica. Esta establece que en todo proceso real existen irreversibilidades —como fricción, disipación y pérdidas térmicas— que degradan la calidad de la energía y limitan el desempeño efectivo de los sistemas.</p>

<p style="text-align:center;">La Primera Ley explica cuánto se conserva ; La Segunda Ley explica por qué no todo se puede convertir en trabajo.</p>

<p style="text-align:center;">El desempeño real es menor que el ideal debido a irreversibilidades explicadas por la Segunda Ley.</p>

# Segunda ley de la termodinámica :

- Primera ley (Cantidad) : hace referencia a la conservación cuantitativa de la energía.
- Segunda ley (Calidad) : hace referencia a posibilidad en la que se puede realizar u proceso. 

## Procesos espontáneos y no espontáneos :
Los procesos espont'aneos son aquellos en los que un sistema para ir de un estado a otro por un proceso, no requiere inyección de energía por parte del medio; de lo contrario, se llamara no espontaneos ya que necesita consumir energía para que un proceso pueda darse sobre ese sistema

## Dirección de proceso :
Ya es importante tener en cuenta en que dirección suceden los procesos, ya que muchas veces, no tener en cuenta este concepto podria incurrirse en procesos imposibles.

## Caso de análisis :



- Tenemos unas bolas de billar, y vamos a golpear la blanca para que colisione con las verdes. Es imposible hacer el inverso, es decir, que las bolas verdes vuelvan al punto inicial, le entreguen la energía con las que fueron golpeadas a la bola blanca y esta le devuelva la energía a la fuente que la golpeo. Es posible volver al estado inicial haciendo un proceso de reorganización, totalmente distinto al proceso del golpe de la bola blanca

---



- En este caso, es imposible que la fuente de baja temperatura que recibio energía de la fuente alta de temperatura, esponteneamente, le devuelva la energía que recibio, es espontano el flujo de fuente alta a fuente baja, pero no es espontaneo el flujo de energía en forma de calor de fuente baja a fuente alta de manera natural

---



- Suponemos un recipiente cerrado en estado estacionario, sin aspas, conectado a un eje y tiene transferencia de calor calor:

Según la primera ley:

$$
\cancel{\Delta E} = \delta Q_{\text{net}} - \delta W
$$

$$
\delta W = \delta Q_{\text{net}}
$$

Pero nunca mueve el eje, por lo que no genera trabajo, solo lo calienta. Se debe diseñar aspas y crear las condiciones iniciales para qeu pueda moverse el eje. Por esto al importancia de tener claro la dirección de los procesos

# Kelvin-Planck / Clausius

## Enunciado de Kelvin-Planck :



- No existen sistemas termodinámicos que convierta el 100% energía en forma de calor en 100% energía de trabajo.
- Todo sistema que convierte parte de Q en W, opera entre fuente de alta temperatura, y fuente baja de temperatura (sumidero donde se rechaza el calor no convertido en trabajo)

## Enunciado Clausius :



- La transferencia de calor de temperatura baja a temperatura alta no es espontaneo, por lo tanto para llevar a cabo este proceso, se requiere consumo de trabajo por parte del sistema.

# Maquinas térmicas

Presentan las siguientes características :

- Se componen de equipos mecánicos.
- Presentan ciclos periódicos.
- Funcionan como ciclos cerrados (no hay ingreso o salida de masa si se toman sus componentes como un solo sistema).
- Funcionan en estado estacionario.
- El fluido de trabajo es aquel que se encarga de recibir y rechazar energía en forma de calor.
- Opera entre temperatura de fuente alta ($T_A$) y temperatura de fuente baja ($T_B$)

Ejemplo: ciclo de vapor o ciclo Rankine :



- Fluido de trabajo : la corriente de agua
- Proceso :
> 1. En la etapa 1, el agua en estado líquido es succionada por una bomba de impulsión, esta consume trabajo.
> 2. En la etapa 2, el agua a aumentado su presión, y entra al intercambiador de calor, donde recibe calor de la fuente $T_A$ para convertirse en estado de vapor sobreaclentado, aumenta su temperatura.
> 3. En la etapa 3, el vapor de agua sobrecalentado entra a una turbina de vapor, donde puede generar trabajo, tiene caida de presión (por lo general no debe de cambiar de fase)
> 4. En la etapa 4, el vapor de agua sobrecalentado entra en un condensador, en el cual se le retira energiá en forma de calor a la fuente $T_B$, bajando el valor de temperatura y por tanto, cambiado de fase a líquido subenfriedo
> 5.  Se repite el ciclo

## Eficiencia Termodinámica : 

Realizamos análisis en sistema cerrado para ciclos termodinámicos :

$$
\Delta E = W_{\text{in}} - W_{\text{out}} + Q_{\text{in}} - Q_{\text{out}}
$$

- $W_{\text{in}} $ : energía que consume la bomba de impulsión
- $W_{\text{out}}$  : energía que genera la turbina
- $Q_{\text{in}} $ : energía que ingresa al sistema en forma de calor desde la fuente T_A   
- $Q_{\text{out}}$  : energía que expulsa sistema en forma de calor hacia la fuente T_B 

- Con estado estacionario, $\Delta E = 0$
- El trabajo neto es : $W_{\text{out}} - W_{\text{in}} = W_{\text{neto}}$

$$
0 = - W_{\text{neto}} + Q_{\text{in}} - Q_{\text{out}} \\
W_{\text{neto}} =  Q_{\text{in}} - Q_{\text{out}} 
$$

Así, vamos a denotar la eficiencia térmica como:

*- El porcentaje de la energía en forma de calor ingresada al sistema, en la que se puede aprovechar como energía en forma de trabajo -*

$$
\checkmark \eta_{\text{termica}} = \frac{W_{\text{neto}}}{Q_{\text{in}}}
$$

O también se puede expresar de la siguiente manera:

$$
\eta_{\text{termica}} = \frac{ Q_{\text{in}} - Q_{\text{out}}}{Q_{\text{in}}} = \frac{ Q_{\text{in}}}{Q_{\text{in}}} - \frac{Q_{\text{out}}}{Q_{\text{in}}}
$$
$$
\checkmark \eta_{\text{termica}} = 1 - \frac{Q_{\text{out}}}{Q_{\text{in}}}
$$

Muchas veces se asume que el trabajo neto, es el trabajo de salida el sistema.

# Maquina de refrigeración :

Objetivo: Mantener una zona de interés a una baja  (respecto al exterior).



Para esta situación ya no hablamos del concepto de *Eficiencia térmica*, ya que una maquina de refrigeración no va a generar trabajo sino consumirlo, entonces hablamos de **Coeficiente de actuación (COP)**, tambien llamado en algunos casos como coeficiente de desempeño:

$$
\checkmark COP_r = \frac{Q_{\text{in}}}{W_{\text{in}}}
$$

Nuestro objetivo es mantener el espacio de interés a una baja temperatura, y la idea es que necesitemos el menor trabajo posible para extraer retirar la mayor cantidad de energía en forma de calor posible de ese espacio, y enviarlo al medio. Por lo tanto se debe asegurar en lo posible para que el sistema valga la pena ejecutar, que $Q_{\text{in}} > W_{\text{in}}$, no es una condición de fuerza, pero si es ideal que el calor que se retire del espacio sea mucho mas grande que el trabajo que se entregue a la maquina

Por primera ley :

$$
\cancel{\Delta E} =  Q_{\text{in}} - Q_{\text{out}} + W_{\text{in}} \\
W_{\text{in}} =  Q_{\text{out}} - Q_{\text{in}} 
$$

$$
\checkmark COP_r = \frac{Q_{\text{in}}}{W_{\text{in}}} =  \frac{Q_{\text{in}}}{Q_{\text{out}} - Q_{\text{in}}} = \frac{1}{\frac{Q_{\text{out}}}{Q_{\text{in}}} - 1}
$$

Recordar que:

- $Q_{\text{in}}$ : es el calor que **RECIBE** a la maquina de refrigeración por parte del espacio a mantener refrigerado, **NO** es el calor que le entra al medio externo.
- $Q_{\text{out}}$ : es el calor que **RECHAZA** la maquina frigorífica y recibe el medio externo, **NO** el que saca de el lugar a refrigerar

siempre ante cualquier situación, nuestro sistema sera la maquina de regrigeración, no el lugar a refrigerar ni el medio que recibe calor por parte de la maquina

# Bomba de calor

Objetivo: Mantener una zona de interés a una alta temperatura (respecto al medio).



De igual manera, seguimos usando el concepto de **Coeficiente de actuación (COP)** para bomba de calor :

$$
\checkmark COP_b = \frac{Q_{\text{out}}}{W_{\text{in}}}
$$

Ahora nuestro objetivo es mantener a una temperatura alta un espacio objetivo, que generalmente es mas alto que las condiciones ambientales externas, por lo tanto, nuestro foco es la cantidad de calor que recibe el espacio a mantener a altas temperaturas. Tambien es ideal que se cumpla que: $Q_{\text{out}} > W_{\text{in}}$ para tener mayor eficiencia en la bomba de calor

$$
\checkmark COP_b = \frac{Q_{\text{out}}}{W_{\text{in}}} =  \frac{Q_{\text{out}}}{Q_{\text{out}} - Q_{\text{in}}} = \frac{1}{1-\frac{Q_{\text{in}}}{Q_{\text{out}}}}
$$

Recordar que:

- $Q_{\text{in}}$ : es el calor que **RECIBE** a la bomba de calor por parte de condiciones ambientales que es mas frio que el espacio objetivo, **NO** es el calor que le entra al espacio objetivo o de interés.
- $Q_{\text{out}}$ : es el calor que **RECHAZA** la bomba de calor y recibe el espacio a tener a temperaturas altas respecto a las condiciones ambientales, **NO** el que saca de el ambiente a condiciones de baja temperatura (relativamente frias)

siempre ante cualquier situación, nuestro sistema sera la bomba de calor, no el lugar en el que subira la temperatura ni el medio amebiente que entrega calor a la bomba de calor.

# Ciclo ideal de carnot



Procesos del ciclo ideal de carnot :

1. $(A \rightarrow B)$ Proceso isotérmico, ingresa calor.
2. $(B \rightarrow C)$ Proceso adiabático, hay cambio de temperatura.
3. $(C \rightarrow D)$ Proceso isotérmico, se rechaza calor.
4. $(B \rightarrow C)$ Proceso adiabático, hay cambio de temperatura.

para el proceso de Carnot ideal, se considera los siguientes aspectos:

- Se considera que el fludo de trabajo es un gas ideal ($Pv = RT$).
- En las operaciones donde se da transferencia de calor, es isotermico.
- En las turbina y el compresor, el proceso es adiabático.
- Las operaciones se dan en estado estacionario en donde hay transferencia de calor.

1. En la operación de combustión: $T_A = T_B = T_1$
$$
\begin{aligned}
\mathrm{d}E &= \delta Q - \delta W \\
\cancel{\mathrm{d}u} &= \delta Q - \delta W \\
\delta Q &= \delta W ; \delta W = P\mathrm{d}v ; P = \frac{RT_1}{v}\\
\delta Q &= RT_1\frac{\mathrm{d}v}{v} \\[6pt]
Q_\text{in} &= RT_1 \ln\left(\frac{v_B}{v_A}\right)\\
\end{aligned}
$$

2. Analogamente para el proceso de intercambio de calor entre C y D:

$$
\begin{aligned}
Q_\text{out} &= -RT_2 \ln\left(\frac{v_D}{v_C}\right)\\
Q_\text{out} &= RT_2 \ln\left(\frac{v_C}{v_D}\right)\\
\end{aligned}
$$

Por lo tanto, hasta ahora, la eficiencia termodinamica del ciclo se da por:

$$
\eta = 1 - \frac{Q_{\text{out}}}{Q_{\text{in}}} = 1 - \frac{RT_2 \ln\left(\frac{v_C}{v_D}\right)}{RT_1 \ln\left(\frac{v_B}{v_A}\right)}\\[6pt]
\eta = 1 - \frac{T_2 \ln\left(\frac{v_C}{v_D}\right)}{T_1 \ln\left(\frac{v_B}{v_A}\right)}
$$

Continuamos con el proceso de la turbina y compresor:

3. En la operación de la turbina: $Q = 0$

$$
\begin{aligned}
\mathrm{d}E &= \cancel{\delta Q} - \delta W \\
\mathrm{d}u &= - \delta W \\
\mathrm{d}u - \delta W &= 0 ; \mathrm{d}u = C_v(T)\mathrm{d}T\\
C_v(T)\mathrm{d}T + P\mathrm{d}v &= 0 ; P = \frac{RT_1}{v}\\
C_v(T)\mathrm{d}T + RT\frac{\mathrm{d}v}{v} &= 0 \\[6pt]
\end{aligned}
$$

En este punto, la ultima ecuación la dividimos toda por el termino $C_v(T) . T$

$$
\begin{aligned}
\frac{\mathrm{d}T}{T} + \frac{R}{C_v(T)}\frac{\mathrm{d}v}{v} &= 0 \\[6pt]
\end{aligned}
$$
ahora con (Recordar relaciones de capacidades caloríficas en el capítulo 4):

$$
\frac{(R = C_p(T) - C_v(T))}{C_v(T)} \\
\frac{R}{C_v(T)} = \frac{C_p(T)}{C_v(T)} - 1 \\
\frac{R}{C_v(T)} = \gamma - 1 \\
$$

Tenemos entonces : 
$$
\begin{aligned}
\int_{T_1}^{T_2} {\frac{\mathrm{d}T}{T}} + (\gamma - 1)\int_{v_B}^{v_C} \frac{\mathrm{d}v}{v} &= 0 \\[6pt]
RT_2 \ln\left(\frac{T_2}{T_1}\right) + (\gamma - 1) \ln\left(\frac{v_C}{v_B}\right) &= 0 \\[6pt]
\frac{T_2}{T_1} &=  \left({\frac{v_C}{v_B}}\right)^{(\gamma - 1)}
\end{aligned}
$$

4. De manera análoga para el compresor : 
$$
\frac{T_1}{T_2} =  \left({\frac{v_A}{v_D}}\right)^{(\gamma - 1)}
$$

por tanto:
$$
\left({\frac{v_D}{v_A}}\right)^{(\gamma - 1)} = \left({\frac{v_C}{v_B}}\right)^{(\gamma - 1)}
$$

lleva a que :
$$
\frac{v_B}{v_A}= \frac{v_C}{v_D}
$$

Volvemos a eficiencia:

$$
\eta = 1 - \frac{T_2 \cancel{\ln\left(\frac{v_C}{v_D}\right)}}{T_1 \cancel{\ln\left(\frac{v_B}{v_A}\right)}} \\[6pt]
$$
$$
\checkmark \eta = 1 - \frac{T_2}{T_1}
$$

Ante toda la demostración anterior, se observa que para condiciones ideales, la eficiencia térmica de un ciclo jamas sera el 100% (la unica manera es que $T_2$ tienda a cero y que $T_1$ tienda a infinito), por lo que para situaciones reales, la eficiencia real esta por debajo de la eficiencia ideal

# Eficiencia térmica ideal de un ciclo:
$$
\checkmark \eta_{\text{ideal}} = 1 - \frac{T_2}{T_1}
$$

Si un ciclo operara entre 850K y 280K, la eficiencia ideal seria $1 - \frac{280}{850} = 67\% $, por tanto, en condiciones reales, esta por debajo de este porcentaje, ya que en la realidad, no es posible que algún proceso térmico este en condiciones ideales.

Para concluir:
- La eficiencia térmica ideal funciona para cuando el flujo de trabajo es un gas ideal.
- En los procesos donde hay flujo de calor, son procesos isotermicos y estacionarios.
- En los procesos donde hay generación/consumo de energía en forma de trabajo, son adiabáticos.
- Funciona para procesos reversibles.
- $\eta_{\text{ideal}}$ es equivalente a $\eta_{\text{reversible}}$
- $\eta_{\text{irreversible}}$ es el que generalmente se mide en procesos **REALES**.

$$
\begin{aligned}
\eta_{\text{reversible}} &> \eta_{\text{irreversible}} \rightarrow \text{Es un proceso irreversible} \\
\eta_{\text{reversible}} &= \eta_{\text{irreversible}} \rightarrow \text{Es un proceso reversible} \\
\eta_{\text{reversible}} &< \eta_{\text{irreversible}} \rightarrow \text{Es un proceso IMPOSIBLE} \\
\end{aligned}
$$
