<p style="text-align:center;">Se realizan análisis de primera ley de la termodinámica para sistemas abiertos (volumenes de control), en el cual, ya hay flujos de masa tanto en la entrada como salida de sistemas a analizar.</p>

# Balance de masa para sistemas abiertos : 



Según la primera ley de la termodinámica, la masa es constante en el universo, por tanto :

$$
\Sigma m_{\text{in}} - \Sigma m_{\text{out}} = \frac{\mathrm{d}m}{\mathrm{d}t}
$$

Lo anterior indica que todas las corrientes de entrada del sistema, menos todas las corrientes de salida del sistema es igual a la acumulación de masa dentro del sistema. Para sistemas estacionarios tenemos que 

$$
\Sigma m_{\text{in}} - \Sigma m_{\text{out}} = \cancel{\frac{\mathrm{d}m}{\mathrm{d}t}}
$$
$$
\Sigma m_{\text{in}} - \Sigma m_{\text{out}} = 0
$$
$$
\Sigma m_{\text{in}} = \Sigma m_{\text{out}} 
$$

## Ecuaciones de continuidad:

$$
\rho = \frac{m}{V} =  \frac{\dot m}{\dot V} \: \: ; \: \: \dot V = A . \vec V
$$
$$
\rho = \frac{\dot m}{A . \vec V} \rightarrow \dot m = \rho . A . \vec V \rightarrow \dot m = \frac{A . \vec V }{v}
$$
Por tanto, aplicando loe balances de masa para estado estacionario :
$$
\Sigma_{\text{in}} \frac{A . \vec V }{v} = \Sigma_{\text{out}} \frac{A . \vec V }{v} 
$$



# Balance de energía para sistemas abiertos :

$$
\Sigma_{\text{in}}m_ie_i - \Sigma_{\text{in}}m_ie_i + Q - W = \frac{\mathrm{d}E}{\mathrm{d}t} 
$$

donde *e* es la contribución de energía en todas sus formas  es :

$$
e_n = u_n + \frac{\vec V^2_n}{2} + gZ_n + P_nv_n
$$

recuerdese que :

$$
h_n = u_n + P_nv_n
$$

por tanto :
$$
e_n = h_n + \frac{\vec V^2_n}{2} + gZ_n
$$

como se puede observar, se tiene en cuenta la energía interna por interacciones moleculares, que es función de temperatura, trabajo mecánico por diferencia de presión que hace que el fluido se mueva, energía cinética y energía potencial


## Nomenclatura para la letra v :

$$
\begin{aligned}
\vec{V} &: \text{vector velocidad} \\ \\
V &: \text{volumen} \\ \\
v &: \text{volumen específico } [=] \frac{\text{volumen}}{\text{masa}} \\ \\
\dot V &: \text{caudal volumétrico } [=] \frac{\text{volumen}}{\text{tiempo}} \\ \\
\overline{v} &: \text{volumen específico molar } [=] \frac{\text{volumen}}{\text{mol}}
\end{aligned}
$$
