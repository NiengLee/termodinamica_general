```python
print("Realizado por ExergiaCode: ")
#   Condiciones inciales - sist A, sist B
V_A = 6                # m
vesp_B = 0.417         # m/kg

#   Condiciones finales - sist T
V_T = 11               # m
vesp_T = 0.55          # m/kg

#   Desarrollo
#   Objetivo : vol_esp A [=] m/kg

V_B = V_T - V_A
m_b = V_B / vesp_B
m_T = V_T / vesp_T 

m_A = m_T - m_b

vesp_A = V_A / m_A

print(f"el volumen especifico en A es = {vesp_A:,.3f} m/kg")
```

# Pagina 29, 34 - Kennet wark

# Pagina 30 , 43 - Kennet wark



```python
print("Realizado por ExergiaCode: ")
# Condiciones
T_amb = 25 + 273.15 # K
P_0 = 1 * (100000) # bar = 100000 Pa
den_0 = 1.19 # kg/m
g = 9.8 # m/s

# Objetivo : determinar P y densidad a) 2000m b) 800m por encima del nivel del mar
# Pv = RT
h_2000 = 2000 #m
h_800 = 800 #m

vesp_0 = 1 / den_0

R = (P_0 * vesp_0)/T_amb

P_2000 = P_0 * math.exp((-g* h_2000)/(R*T_amb) )
P_800 = P_0 * math.exp((-g * h_800)/(R*T_amb))

den_2000 = P_2000 * (den_0 / P_0)
den_800 = P_800 * (den_0 / P_0)

print(f"los valores de presion y volumen para h = {h_2000} m son respectivamente {P_2000/100000:,.3f} bar y {den_2000:,.3f} kg/m")
print(f"los valores de presion y volumen para h = {h_800} m son respectivamente {P_800/100000:,.3f} bar y {den_800:,.3f} kg/m")
```
