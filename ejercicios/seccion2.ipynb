# Pagina 82 , 74 - Kennet wark



```python
print("Realizado por ExergiaCode : ")

#   Condiciones iniciales : 
Vi = 0.12 # m

#   Condiciones finales : 
Vf = 0.04 # m

#   Desarrollo : Pv = RT [1 + (b/v)], b = 0.012 m/kg, R = 8.314 kJ/kmol.K
m = 1 # kg
PM = 60 # kg/kmol
T = 27 + 273.15 # K
F_fr = 10000 # N
x = 0.5 # m
P_atm = 100 * (1000) # Pa
R = 8.314 * (1000) * (1/PM) # J/kg.K
b = 0.012 # m/kg

#   Objetivos : determinar W_comp con friccion ; W_comp sin friccion [=] N.m = J
vi = Vi/m
vf = Vf/m

v = sp.Symbol("v")
expresion = -((R*T)*(1+(b/v)))/(v)
W_resultado_int = sp.integrate(expresion, (v,vi,vf))

print(f"el trabajo de compresion es {W_resultado_int:,.3f} J")

# b)
expresion_atm = P_atm
W_result_atm = sp.integrate(expresion_atm, (v,vi,vf))

W_Ffr = -F_fr * x
W_total = W_resultado_int + W_result_atm - W_Ffr

print(f"el trabajo total es {W_total:,.3f} J")
```

# Pgina 102 , 74 - Cengel



```python
print("Realizado por ExergiaCode : ")

#   Condiciones iniciales
D_in = 8 * (1/100) # m
r_in = D_in/2
A_in = np.pi * (r_in**2)
#   Condiciones finales
D_out = 12 * (1/100) # m
r_out = D_out/2
A_out = np.pi * (r_out**2)

#   Desarrollo
W_in = 44 * (1000) # Watt = J/s
caudal = 0.1 # m/s
den_ac = 860 # kg/m
delta_P = 500 * (1000) # Pa
eficiencia_motor = 0.9

#   Objetivo : determinar la eficiencia mecnica de la bomba
m = caudal * den_ac
vel_out = caudal / A_out
vel_in = caudal / A_in
contrib_Ec = ((vel_out**2)-(vel_in**2)) / 2
contrib_P = delta_P / den_ac

W_bomba = m * ( contrib_Ec + contrib_P)

print(f"la energa que consume la bomba para realizar el trabajo mecnico es {W_bomba/1000:,.3f} kJ")

W_eje_real = eficiencia_motor * W_in

print(f"El trabajo real generado por el motor es :  {W_eje_real/1000:,.3f} kJ")

print(f"el porcentaje de energa consumida por la bomba es :  {(W_bomba /W_eje_real)*100:,.3f} %")
```
