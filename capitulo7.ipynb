# Desigualdad de Clausius




Denotaremos por convecion que $Q>0$ como el calor que entra al sistema y tambien que $W>0$ trabajo que sale del sistema y además:
- Tenemos un sistema cerrado S.
- Tenemos un dispositivo cíclico reversible.
- Una única fuente térmica a temperatura constante.

Empezamos con eficiencia térmica de un ciclo de Carnot **IDEAL** : 

$$
\eta_{\text{ideal}} = 1 - \frac{T}{T_A}
$$

Que aplica para ciclos reversibles y las temperaturas $T_A$ y $T$ son constante. Tenemos eficiencia térmica :

$$
\eta_{\text{termica}} = 1 - \frac{Q}{Q_{\text{A}}}
$$

si tuvieramos un ciclo ideal, sería lógico que:

$$
1 - \frac{T}{T_\text{A}} = \eta_{\text{termica, ideal}} = 1 - \frac{Q}{Q_{\text{A}}}
$$

Para dispositivos reversibles:

$$
\frac{T}{T_\text{A}} = \frac{\delta Q}{\delta Q_{\text{A}}} \\[10pt]
\frac{\delta Q_{\text{A}}}{T_\text{A}} = \frac{\delta Q}{T}\\[10pt]
\delta Q_{\text{A}} = {T_\text{A}} * \frac{\delta Q}{T}\\
$$

Ahora, el sistema compuesto opera en ciclo y solo intercambia calor con una única fuente $T_A$. Por Kelvin–Planck: no puede producir trabajo neto,es decir:

- 100% de $Q_A$ $\cancel{\rightarrow}$ 100% en forma de W.
- El sistema completo que recibe calor no rechaza calor al medio.
- No puede cumplirse $W_{\text{total}} > 0$, Entonces...

$$
W_{\text{total}} \leq 0
$$

y por primera ley de la termodinámica :

$$
\cancel{\mathrm{d}E} = \delta Q_A - \delta W\\
\oint \delta Q_A = W_{\text{total}}\\
$$

con lo anterior se concluye que $\oint \delta Q_A = W_{\text{total}} \leq 0$

Si todas las temperaturas estan en escala absoluta, y $\oint \delta Q_A \leq 0$, entonces para 

$$
\delta Q_{\text{A}} = {T_\text{A}} * \frac{\delta Q}{T}\\[10pt]
\oint{\delta Q_{\text{A}}} = {T_\text{A}} * \oint\frac{\delta Q}{T}\\[10pt]
\oint\frac{\delta Q}{T} \leq 0 \rightarrow \text{Esta es la desigualdad de Clausius}
$$

# Deficinición formal de entropía : 

La desigualdad :

$$
\oint\frac{\delta Q}{T} \leq 0
$$

Es una propiedad universal de todo ciclo real, derivada únicamente de la segunda ley, además, 

- Se cumple para cualquier sustancia
- Se cumple para cualquier ciclo
- No depende de ecuaciones de estado
- No depende del tipo de proceso
- Solo depende de la segunda ley

Eso significa que la desigualdad tiene un comportamiento universal:

- Vale 0 si el ciclo es reversible
- Es menor que 0 si es irreversible

Esta universalidad permite definir una nueva función de estado S tal que:

$$
\mathrm{d}S = \frac{\delta Q_{\text{rev}}}{T}
$$

La segunda ley impone una restricción universal sobre cualquier ciclo posible.

### Expresiones a tener en cuenta:

$$
\begin{align}
\mathrm{d}S &= \frac{\delta Q_{\text{rev}}}{T} \rightarrow \delta Q_{\text{rev}} = T\mathrm{d}S \: ; \: \text{Para procesos internamente reversibles}\\
\mathrm{d}S &> \frac{\delta Q_{\text{rev}}}{T} \rightarrow \delta Q_{\text{rev}} < T\mathrm{d}S \: ; \: \text{Para procesos internamente irreversibles}\\
\end{align}
$$

Finalmente:

$$
\Delta S = S_2 - S_1 = \int_1^2 \frac{\delta Q_{\text{rev}}}{T} 
$$

# Postulado de la segunda ley: 

>- Existe una propiedad íntrinseca y extensiva llamada entropía **S**.
>- La entropía transporte se transporta por transferencia de calor a través de las fronteras de un sistema cerrado. El flujo de entropía asociado con la transferencia d ecalor en una frontera con flujo de calor $\dot{Q}$ y a temperatura uniforme T se define como la magnitud $\frac{\dot{Q}}{T}$, y esta magnitud tiene el mismo convenio de signo que el del flujo de calor.
>- La entropía sólo puede ser producida, y en el límite de proceso internamente reversible la producción de entropía se reduce a cero. Así, la producción de entrpía por unidad de tiempo $\dot{\sigma}$ para un sistema cualquiera debe satisfacer:
>$$
>\dot{\sigma} > 0 \rightarrow \text{Para procesos internamente irreversibles}\\
>\dot{\sigma} = 0 \rightarrow \text{Para procesos internamente reversibles}
>$$

Por tanto para procesos reales:

$$
\begin{aligned}
\oint \left(\frac{\delta Q}{T}\right)_{\text{real}} &= \oint \left(\frac{\delta Q}{T}\right)_{\text{reversible}} + \oint \left(\frac{\delta Q}{T}\right)_{\text{irreversible}}\\[10pt]
\text{Propiedad real} &= \text{Componente reversible del proceso} + \text{Componente irreversible del proceso}
\end{aligned}
$$

Así, para sistemas cerrados :

$$
\begin{aligned}
\mathrm{d}{S_{\text{sistema cerrado}}} = \frac{\delta Q}{T} + \delta \sigma \\[10pt]
S_2 - S_2 = \int \frac{\delta Q}{T} + \sigma
\end{aligned}
$$

En este caso, $\sigma$ no es una propiedad de estado, es una función de proceso, ya que dependiendo del proceso que se siga, el valor de esta irreversibilidad es diferente y tiene todo el sentido (varia con el material si es conductivo o no, presencia de fricción, temperatura ambiental, etc...), para procesos reversibles, este es igual a 0

Para una comprensión mas aproximada de la entropía, esta se refiere al calor que no se pudo generar como trabajo util, por lo tanto se puede definir como el calor perdido por frontera del sistema, sobre el valor de temperatura de la superficie por donde escapa esa energía en forma de calor

### Análisis de importancia!:

Para un sistema adiabático cerrado, tenemos que la transferencia de calor $Q = 0$, así:

$$
\begin{aligned}
\mathrm{d}{S_{\text{sistema cerrado}}} &\geq \frac{{\cancel{\delta Q}}}{T} + \delta \sigma \\[10pt]
\mathrm{d}{S_{\text{sistema cerrado}}} &\geq 0\\[10pt]
S_{\text{final}} - S_{\text{inicial}} &\geq 0\\[10pt]
S_{\text{final}} &\geq S_{\text{inicial}} 
\end{aligned}
$$

Con lo anterior es importante tener presente, que la entropía en el análisis de un sistema tiende a aumentar, es decir, la energía contenida tiende a degradarse y perder su capacidad de generar trabajo util



# Detalles conceptuales :

- $\oint \frac{\delta Q}{T} \leq 0 \rightarrow$ se reconoce la desigualdad de Clausius, T es la temperatura de superficie por donde fluye Q.
- $\mathrm{d}s_{\text{reversible}} = \frac{\delta Q}{T} \rightarrow$ se da para procesos reversibles.
- $\mathrm{d}s_{\text{irreversible}} = \mathrm{d}s_{\text{reversible}} + \sigma \rightarrow$ se da para procesos irreversibles, donde $\sigma$ es una propiedad función de proceso.
- Naturalmente, los procesos siempre se dan de manera espontanea de un estado de entropía baja a entropía alta.
- $S_2 \geq \int_1^2 \frac{\delta Q}{T} + S_1$
- Nunca se presentara en condiciones irreversibles, o reales que de manera espontanea $\Delta S < 0$ y $\sigma < 0$

# Análisis microscópico de la entropía :

A pesar de que en esta etapa no se realiza análisis de la termodinámica estadística, vale la pena interpretar a nivel molecular como se interpreta la entripía. *S* es la aleatoriedad de la distribución molecular, *S* aumenta a medida que no se pueda medir con precisión la posición de las partículas en un sistema a nivel molecular en un estado dado.

$$
S = K \ln \Sigma W
$$

- $K = \text{es la constante de boltzmann }  (1.18*10^{-23} \frac{J}{K})$
- $W_i = \frac{n!}{n_1!n_2!...n_m!}$ indica todas las configuraciones posibles para el estado i, de sus moléculas internas.

# Ecuación de Gibbs para determinar la entropía.

Permite determinar el cambio de entropía de un sistema en función de otras propiedades como presión o volumen, ya que muchas veces medir esta expresión $\frac{\delta Q}{T}$ no es sencillo



- El cambio de entropía en el sistema A pierde energía en forma de calor, sus moleculas ya no se mueven de manera tan caótica, su entropía interna baja.
- El cambio de entropía en el sistema B recibe energía en forma de calor de A, sus moleculas aumentan su energía, su entropía interna aumenta.
- $\Delta S_B = \Delta S_A + \sigma$
- En procesos reversibles el cambio de entropía en el sistema B, es mayor que el cambio de entropía en el sistema A.

Ahora:

- $\mathrm{d}E = \delta W + \delta Q$ es la primera ley de la termodinámica para sistemas cerrados.
- $Pv = RT$ es la ecuación de estado de gas ideal.
- $\delta W = -P\mathrm{d}v$ es el trabajo de frontera.
- $\mathrm{d}u = C_v(T)\mathrm{d}T$ cambio de energía interna en función de capacidad calorífica a volumen constante.
- $\mathrm{d}h = C_p(T)\mathrm{d}T$ cambio de entalpía en función de capacidad calorífica a presión constante.
- $\delta Q = T\mathrm{d}s$ reordenando al segunda ley de la termodinámica.

Ignoramos cambios de energía cinética y potencial:

$$
\mathrm{d}u = \delta W + \delta Q
$$

Reemplazamos delta de W y Q:

$$
\begin{aligned}
\mathrm{d}u &= -P\mathrm{d}v + T\mathrm{d}s \\[10pt]
T\mathrm{d}s &= \mathrm{d}u + P\mathrm{d}v  \rightarrow \text{Es la ecuación de Gibbs}\\[10pt]
\mathrm{d}s &= \frac{\mathrm{d}u}{T} + \frac{P\mathrm{d}v}{T} \: ; \: \text{Por gas ideal:} \: \frac{P}{T} = \frac{R}{v}\\[10pt]
\checkmark \mathrm{d}s &= \frac{ C_v(T)\mathrm{d}T}{T} + \frac{R\mathrm{d}v}{v}
\end{aligned}
$$

Ya tenemos una expresión para calcular cambio de entropía en sistemas cerrados, la capacidad calorífica fue tratado en el capítulo 4, y la segunda expresión es una integral sobre el voluman que tiene como resultado un logaritmo natural. Ahora, entapía:

$$
\begin{aligned}
h &= u + Pv \rightarrow \text{el proceso ya no se da a presión constante} \\[10pt]
\mathrm{d}h &= \mathrm{d}u + \mathrm{d}(Pv)\\[10pt]
\mathrm{d}h &= \mathrm{d}u + v\mathrm{d}P + P\mathrm{d}v\\[10pt]
\mathrm{d}h - v\mathrm{d}P &= \mathrm{d}u + P\mathrm{d}v \rightarrow \text{la expresión despues de la igualdad es Tds}\\[10pt]
\mathrm{d}h - v\mathrm{d}P &= T\mathrm{d}s \\[10pt]
T\mathrm{d}s &= C_p(T)\mathrm{d}T - v\mathrm{d}P   \\[10pt]
\mathrm{d}s &= \frac{C_p(T)\mathrm{d}T}{T} - \frac{v\mathrm{d}P}{T}\: ; \: \text{Por gas ideal:} \: \frac{v}{T} = \frac{R}{P}\\[10pt]
\checkmark \mathrm{d}s &= \frac{C_p(T)\mathrm{d}T}{T} - \frac{R\mathrm{d}P}{P}
\end{aligned}
$$

Ya tenemos una expresión para calcular cambio de entropía en sistemas abiertos.

> Recuerde que lo anterior funciona, para Gas ideal, condiciones ideales!!.

# Resumen para el cálculo de entropía para gases y líquidos

## Gases:
Usamos las mismas expresiones deducidas:

- $\mathrm{d}s = \frac{ C_v(T)\mathrm{d}T}{T} + \frac{R\mathrm{d}v}{v}$ para procesos a volumen constante (sistemas cerrados).
- $\mathrm{d}s = \frac{C_p(T)\mathrm{d}T}{T} - \frac{R\mathrm{d}P}{P}$ para procesos a presión constante (sistemas abiertos).

## Líquidos:
En los líquidos no se presenta cambio de volumen específico de manera notable (son fluidos incompresibles).

$$
\mathrm{d}s = \frac{ C_v(T)\mathrm{d}T}{T}
$$

En caso de tener información tabulada, funciona bien hacer aproximación de las condiciones del sistema líquido al estado saturado.

## Solución de cambio de entropía en gases
Vamos a resolver la integral de cambio de entropía : 

$$
\mathrm{d}s = \frac{C_p(T)\mathrm{d}T}{T} - \frac{R\mathrm{d}P}{P}
$$

Resultado:

$$
\Delta s = \int_{T_1}^{T_2} \frac{C_p(T)\mathrm{d}T}{T} - R\ln{\frac{P_2}{P_1}}
$$

Muchas veces no se dispone de facilidad para calcular la integral desde el estado uno al estado dos del primer termino del lado derecho de la ecuación, por tanto se recurre a lo siguiente:

$$
\begin{aligned}
S_1^0 &= \int_{T_0=0K}^{T_1} \frac{C_p(T)\mathrm{d}T}{T}\\[10pt]
S_2^0 &= \int_{T_0=0K}^{T_2} \frac{C_p(T)\mathrm{d}T}{T}\\[10pt]
\int_{T_1}^{T_2} \frac{C_p(T)\mathrm{d}T}{T} &= \int_{T_0=0K}^{T_2} \frac{C_p(T)\mathrm{d}T}{T} - \int_{T_0=0K}^{T_1} \frac{C_p(T)\mathrm{d}T}{T}\\[10pt]
\int_{T_1}^{T_2} \frac{C_p(T)\mathrm{d}T}{T} &= S_2^0 - S_1^0\\[10pt]
\end{aligned}
$$

Finalmente:

$$
\checkmark \Delta s = S_2^0 - S_1^0 - R\ln{\frac{P_2}{P_1}}
$$

muchas veces la condición $S_1^0$ no necesariamente es en el 0 absoluto de Temperatura, pueden ser condiciones atmosféricas de T = 25°C y P = 1atm

# Analisis de cambio neto de entropía

Como vimos en la expresión anterior vamos a analizar varias situaciones. En muchas situaciones, se indicara analizar sistemas adiabáticos internamente reversibles, esto es un comportamiento isoentrópico, es decir que $\Delta s = 0$, se recomienda también tener presente que si se indica un proceso como isoentrópico, se puede asumir que el proceso es adiabático como aproximación.

## Presión relativa:

Evitar la confusión con presión reducida ($P_{\text{reducida}} = \frac{P_{\text{gas}}}{P_\text{crítica}}$).

Analizamos un proceso isoentrópico:

$$
\begin{aligned}
\cancel{\Delta s} &= S_2^0 - S_1^0 - R\ln{\frac{P_2}{P_1}}\\[10pt]
S_2^0 - S_1^0 &= R\ln{\frac{P_2}{P_1}}\\[10pt]
\ln{\frac{P_2}{P_1}} &= \frac{(S_2^0 - S_1^0)}{R} \\[10pt]
\frac{P_2}{P_1} &= \mathrm{e}^\frac{(S_2^0 - S_1^0)}{R} \\[10pt]
\frac{P_2}{P_1} &= \frac {\mathrm{e}^{\frac{S_2^0}{R}}}{\mathrm{e}^{\frac{S_1^0}{R}}} \\[10pt]
\end{aligned}
$$

Se define entonces la presión relativa como $P = \mathrm{e}^{\frac{S^0}{R}}$, se puede aplicar entonces para:

- Estado de gas
- Procesos insoentrópicos

# Entropía en eficiencias mecánicas

Generalmente se realizan análisis a dispositivos consumen o generan trabajo útil, estos son:

- Compresores.
- Turbinas.
- Bomba de impulsión.
- Tobera.

Tengase en cuenta que los en los procesos ideales se dan de manera isoentrópica.

En este caso, tenemos presente que la eficiencia térmica de un sistema se da por $\eta_{\text{térmica}} = \frac{W_{\text{generado}}}{Q\text{in}}$, esto indica, que porcentaje de la energía en forma de calor que ingresa al sistema se conviernte en trabajo util. Pero las eficiencias que veremos acontinuación, lo que se evidencia en cuanto a la generación de energía (o consumo) en procesos irreversibles, versus procesos isoentrópicos:

## Turbina:
Su objetivo es generar trabajo util, por lo tanto:

$$
\eta_{\text{Turbina}} = \frac{W_{\text{real}}}{W_{\text{ideal}}}= \frac{h_{\text{in}} - h_{\text{out}}}{h_{\text{in}} - h_{\text{out,s}}}
$$

- $h_{\text{out,s}}$ indica entalpía a entropía constante, es decir, el mismo valor de entropía que en la corriente de entrada.

## Compresor y bomba de impulsión:
El objetivo del compresor y bomba es aumentar la presión un gas, consumiendo trabajo útil, por lo tanto:

$$
\eta_{\text{Compresor}} = \frac{W_{\text{ideal}}}{W_{\text{Real}}}= \frac{h_{\text{out,s}} - h_{\text{in}}}{h_{\text{out}} - h_{\text{in}}}
$$

Es lógico pensar que el consumo de energía en procesos irreversibles es mayor que en procesos isoentrópicos.

## Tobera:
Su objetivo es reducir el area de flujo, aumentando presión y velocidad de flujo:

$$
\eta_{\text{Tobera}} = \frac{E_{\text{cinética real}}}{E_{\text{cinética ideal}}}= \frac{h_{\text{in}} - h_{\text{out}}}{h_{\text{in}} - h_{\text{out,s}}}
$$
