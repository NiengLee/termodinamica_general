# Trabajo y producción de entropía

Recordamos el balance de entropía de un sistema abierto:

$$
\frac{\mathrm{d}S_{SA}}{\mathrm{d}t} = \Sigma_{in} \: s_{in}\dot{m}_{in} - \Sigma_{out} \: s_{out}\dot{m}_{out} + \Sigma_{j=1}^n \frac{\dot{Q}_j}{T_j} + \dot{\sigma}_{SA}
$$

Ahora el balance de energía de un sistema abierto :

$$
\frac{\mathrm{d}E_{SA}}{\mathrm{d}t} = \dot{Q} - \dot{W} + \Sigma_{in} \: \left( h + \frac{\vec{V}}{2} + gz\right)_{in}\dot{m}_{in} - \Sigma_{out} \: \left( h + \frac{\vec{V}}{2} + gz\right)_{out}\dot{m}_{out}
$$

- $\dot{W}$ representa todas las posibles formas de potencia transferida.

##  Trabajo útil comunicado $\dot{W}$ :
Es el trabajo total comunicado, sustrayendo el trabajo realizado por la atmósfera. Sea tambien $P_0$ la presión atmosférica o de presión de las condiciones ambientales y es un valor constante

el trabajo realizado por la atmósfera sobre el sistema es $-P_0 \mathrm{d}V$

$$
\delta W_u = \delta W - (-P_0 \mathrm{d}V)
$$

También se expresa como:

$$
\begin{aligned}
W_u &= W + P_0 \Delta V \\[10pt]
\dot{W}_u &= \dot{W} + P_0 \frac{\mathrm{d}V_{SA}}{\mathrm{d}t} \\[10pt]
\dot{W} &= \dot{W}_u + P_0 \frac{\mathrm{d}V_{SA}}{\mathrm{d}t} \\[10pt]
\end{aligned}
$$

Ahora se toma la ecuación de balance de entropía y se multiplica por la temperatura ambiente, condiciones atmosféricas que estan alrededor del sistema que se analiza:

$$
T_0 * \left(\frac{\mathrm{d}S_{SA}}{\mathrm{d}t} = \Sigma_{in} \: s_{in}\dot{m}_{in} - \Sigma_{out} \: s_{out}\dot{m}_{out} + \Sigma_{j=1}^n \frac{\dot{Q}_j}{T_j} + \dot{\sigma}_{SA}\right)
$$

el balance de entropía ya queda en unidades de energía, entonces se **RESTAN** las ecuaciones de balance de primera ley y segunda ley, y reemplazamos $\dot{W}$ en terminos de $\dot{W}_u$

$$
\dot{W}_u = \Sigma_{out} \: \left( h + \frac{\vec{V}}{2} + gz - T_os\right)_{out}\dot{m}_{out} - \Sigma_{in} \: \left( h + \frac{\vec{V}}{2} + gz\right)_{in}\dot{m}_{in} - \Sigma_{j=1}^n Q_J\left(1 - \frac{T_0}{T_j}\right) + \frac{\mathrm{d}\left(E + P_0V - T_0s \right)_{SA}}{\mathrm{d}t} + T_0\dot{\sigma}_{SA}
$$

- La ecuación anterior relaciona directamente la potencia neta útil comunicada a un sistema con la velocidad de producción de entropía en el sistema.
- Para procesos reversibles, en la ecuación anterior $\dot{\sigma}_{SA} = 0$ 
- Se puede asumir que la sumatoria de las temperaturas en cada punto de la superficie del sistema es $T_f$, una temperatura uniforme

# Exergía

Se reconoce como la energía máxima convertible en trabajo útil respecto a un ambiente de referencia, y depende de las condiciones ambientales de $T_0$ y $P_0$.

## Estado muerto
Es aquel estado en el que un sistema esta en equilibrio con las condiciones ambientales/atmosféricas, y al estar en equilibrio mutuo, no es espontáneo el cambio de estado del sistema respecto al medio, por no existir esta diferencia en sus propiedades de temperatura ni presión, ya el sistema no puede realizar trabajo (*Trabajo reversible maximo o potencial de trabajo*)

### Atmosfera estandar:

- $T_0$ = 298.15 K
- $P_0$ = 1 atm ; 1.01325 bar

### IMPORTANTE:

La transferencia de energía asociada a la interacción de trabajo (excluyendo trabajo contra el ambiente) tiene el mismo valor que el propio trabajo útil.

# Exergía en sistemas cerrados

Tenemos la siguiente expresión para trabajo util:

$$
\dot{W}_u = \Sigma_{out} \: \left( h + \frac{\vec{V}}{2} + gz - T_os\right)_{out}\cancel{\dot{m}_{out}} - \Sigma_{in} \: \left( h + \frac{\vec{V}}{2} + gz\right)_{in}\cancel{\dot{m}_{in}} - \Sigma_{j=1}^n Q_J\left(1 - \frac{T_0}{T_j}\right) + \frac{\mathrm{d}\left(E + P_0V - T_0s \right)_{SA}}{\mathrm{d}t} + T_0\cancel{\dot{\sigma}_{SA}}
$$

- En sistemas cerrados no hay ingreso ni salida de masa
- Asumimos que el trabajo tiene proceso reversible util, por lo tanto $\sigma = 0$

$$
\begin{aligned}
\dot{W}_{rev,u} &= \frac{\mathrm{d}\left(E + P_0V - T_0S \right)_{SC}}{\mathrm{d}t} - \Sigma_{j=1}^n Q_J\left(1 - \frac{T_0}{T_j}\right) \\[10pt]
\delta W_{rev,u} &= \mathrm{d}\left(E + P_0V - T_0S \right)_{SC} - \Sigma_{j=1}^n \left(1 - \frac{T_0}{T_j}\right) \delta Q_J \\[10pt]
W_{rev,u} &= \Delta E + P_0 \Delta V - T_0 \Delta S  - \int_1^2 \left(1 - \frac{T_0}{T_f}\right) \delta Q_J \\[10pt]
\end{aligned}
$$

- Se reemplaza $T_j$ por $T_f$, que es la temperatura uniforme sobre la superfice del sistema
- Evaluaremos el potencial de trabajo sin tener en cuenta transferencia de calor, $T_f$ tendrá el valor $T_0$ ya que se asume que solo tiene interacción con el medio en condiciones ambientales, así $\frac{T_0}{T_f} = 1$ y por tanto:

$$
\begin{aligned}
W_{rev,u} &= \Delta E + P_0 \Delta V - T_0 \Delta S  - \cancel{\int_1^2 \left(1 - \frac{T_0}{T_f}\right) \delta Q_J} \\[10pt]
W_{rev,u} &= E_2 - E_1 + P_0(V_2 - V_1) - T_0(S_2 - S_1) \\[10pt]
\end{aligned}
$$

El estado 2 , es el estado muerto, por tanto, la energía final es la energía interna en el estado muerto, ya que no tiene velocidad cinética ni potencial, y las condiciones inciales no sobreentiende que es sub 1:

$$
W_{rev,u} = U_0 - E + P_0(V_0 - V) - T_0(S_0 - S) \\[10pt]
$$

Ahora la expresión anterior es para el trabajo consumido, pero si es el trabajo generado por el sistema, se invierten los signos, se obtieine finalmente:

$$
\begin{aligned}
W_{rev,out,u} &= E - U_0 + P_0(V - V_0) - T_0(S - S_0) \\[10pt]
\end{aligned}
$$

Por tanto, se define **Exergía** para sistemas cerrados, sin transferencia de calor, y bajo condiciones reversibles como :

$$
\begin{aligned}
\Phi &= E - U_0 + P_0(V - V_0) - T_0(S - S_0) \\[10pt]
\Phi &= (E + P_0V - T_0S) - (U_0 + P_0V_0 - T_0S_0) \\[10pt]
\phi &= \frac{\Phi}{m}\\[10pt]
\end{aligned}
$$

## Exergía asociada a la transferencia de calor : 

Recordamos que:

$$
W_{pot} = Q\eta_{carnot} = Q\left( 1 - \frac{T_0}{T_f}\right)
$$

como lo anterior, es un potencial de trabajo del calor relativo a el estado muerto, esto tambien es exergía, por tanto:

$$
\Phi_{Q,f} = Q\left( 1 - \frac{T_0}{T_f}\right)
$$

La anterior expresión se reconoce como la exergía asociada al calor transferido Q, la expresión anterior puede expresarse como:

$$
\begin{aligned}
\Phi_{Q,f} &= Q - T_0\left(\frac{Q}{T_f}\right) \\[10pt]
\Phi_{Q,f} &= Q - T_0 \Delta S_f \\[10pt]
\end{aligned}
$$

en forma de integral :

$$
\Phi_Q = \int_1^2 \left( 1 - \frac{T_0}{T_j} \right) \delta Q
$$

## Balance de Exergía completo para sistemas cerrados :

El estado final siempre es el estado muerto, es decir, aquel estado en el que el sistema estará en equilibrio termodinámico con las condiciones ambientales

$$
\begin{aligned}
\checkmark W_u &= \Delta U + P_0\Delta V - T_0\Delta S - \int_1^2 \left( 1 - \frac{T_0}{T_j} \right) \delta Q + T_0\sigma_{SC} \\[10pt]
\rightarrow \Delta \Phi_{SC} &= \Delta U + P_0\Delta V - T_0\Delta S \\[10pt]
\rightarrow \Phi_{Q} &= \int_1^2 \left( 1 - \frac{T_0}{T_j} \right) \delta Q \\[10pt]
\rightarrow I_{SC} &= T_0\sigma_{SC} \\[10pt]
W_u &= \Delta \Phi_{SC} - \Phi_{Q} + I_{SC} \\[10pt]
\checkmark \Delta \Phi_{SC} &= \Phi_{Q} + W_u - I_{SC}
\end{aligned}
$$

- Para procesos internamente reversibles, $I_{SC} = 0$
- Para la irreversibilidad $I_{SC}$ que se relaciona solo con transferencia de calor en sistemas cerrados :

$$
I_{Q} = \Phi_{in} - \Phi_{out}
$$

# Exergía en sistemas abiertos :

## En régimen estacionario

Tomamos la siguiente ecuación previamente planteada:

$$
\dot{W}_{u,est} = \Sigma_{out} \: \left( h + \frac{\vec{V}}{2} + gz - T_0s\right)_{out}\dot{m}_{out} - \Sigma_{in} \: \left( h + \frac{\vec{V}}{2} + gz\right)_{in}\dot{m}_{in} - \Sigma_{j=1}^n Q_J\left(1 - \frac{T_0}{T_j}\right) + \cancel{\frac{\mathrm{d}\left(E + P_0V - T_0s \right)_{SA}}{\mathrm{d}t}} + T_0\dot{\sigma}_{SA}
$$

La expresión anterior proporciona le máximo trabajo obtenido o mínimo suministrado por un fluido en un sistame abierto.

### Exergía de una corriente o un flujo:
para un una corriente de un fluido en régimen estacionario, la exergía de corriente se define como el trabajo máximo que puede obtenerse cuando el fluido cambia reversiblemente desde un estado hasta el estado muerto, en un proceso en el que se intercambia calor solo con la atmósfera.

Definimos una nueva propiedad \psi como exergía de una corriente :

$$
\psi = \left( h + \frac{\vec{V}}{2} + gz - T_0s\right) - \left( h_{0} + \cancel{\frac{\vec{V}}{2}} + gz - T_0s_0\right)
$$

- La energía cinética en el estado final es cero, por tanto

$$
\begin{aligned}
\psi &= \Delta h - T_0 \Delta s + \frac{\vec{V}_{inicial}}{2} + g\Delta z \\[10pt]
\Psi &= \frac{\psi}{m}
\end{aligned}
$$

En el cambio de propiedades, se debe tener presente, que el estado final es el estado muerto, entonces se debe buscar el cambio de condiciones iniciales hasta las condiciones de estado muerto ($h_0$, $s_0$, $z_0$)

Así :

$$
\dot{W}_{u,est} = \Sigma_{out} \: \psi_{out}\dot{m}_{out} - \Sigma_{in} \: \psi_{in}\dot{m}_{in} - \Sigma_{j=1}^n Q_J\left(1 - \frac{T_0}{T_j}\right)
$$

Ahora, el trabajo real util es:

$$
\dot{W}_{real,u} = \Sigma_{out} \: \psi_{out}\dot{m}_{out} - \Sigma_{in} \: \psi_{in}\dot{m}_{in} - \Sigma_{j=1}^n Q_J\left(1 - \frac{T_0}{T_j}\right) + T_0\dot{\sigma}_{SA}
$$

Por tanto, de manera análoga a sistemas cerrados, se presenta el balance de Exergía para un sistema abierto en régimen estacionario : 

$$
\begin{aligned}
\checkmark \Sigma_{out} \: \psi_{out}\dot{m}_{out} - \Sigma_{in} \: \psi_{in}\dot{m}_{in} = \Phi_{Q} + \dot{W}_{real,u} - I_{SA}
\end{aligned}
$$

# Efectividad o rendimiento exergético :

$$
\begin{aligned}
\epsilon = \eta_{II} &= \frac{\text{exergía obtenida}}{\text{exergía suministrada}} \\[10pt]
                     &= 1 - \frac{\text{destrucción / perdidas de exergía}}{\text{exergía suministrada}}
\end{aligned}
$$

Ahora para un ciclo de potencia tenemos :

$$
\begin{aligned}
\epsilon_{pot} &= \frac{W_{real}}{W_{reversible}} \\[10pt]
               &= \frac{\eta_{termico,real}}{\eta_{carnot}} \\[10pt]
               &= \frac{\eta_{termico,real}}{ 1 - \frac{T_B}{T_A}} \\[10pt]
\end{aligned}
$$
