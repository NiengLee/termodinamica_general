# Capítulo 01 - Conceptos básicos, definiciones y glosario.

# Termódinamica :
Ciencia que estudia la materia en cuanto a sus transformaciones a nivel macro y micro, como cambian sus propiedades y como afectan sus estados.
Tipos de Termodinámica:

- ## Clásica :
Se encarga de estudiar los sistemas a nivel macroescala.
- ## Estadística
Realiza análisis molecular, simulaciones moleculares, dinamica molecular teniendo en cuenta modelos probabilisticos a nivel microescala y de aquí inferir propiedades macroescala.

# Terminología:
- ## Materia
O masa, y es de lo que esta compuesto un sistema o un cuerpo (átomos, moléculas, biomoléculas, etc), no depende del espacio donde esta ubicado.
(Recordar que peso es igual a masa por gravedad, y el peso si depende de el espacio donde esta sometido a una fuerza de gravedad de atracción entre cuerpos)

- ## Energía
Propiedad almacenada en un sistema para realizar algún tipo de cambio (posición, traslación, rotación, interno)

- ## Transferencia
Mecanismo en el cual dos sistemas intercambian alguna propiedad que ambos puedan compartir

- ## Sistema (Termodinámico)
Región en el espacio en el cual se le desea realizar un analisis de sus propiedades y cambios





- ### Sistema cerrado (masa de control)
Sistema en el que no tiene intercambio de masa con el medio, solo tiene interacciones energéticas .

- ### Sistema abierto (volumen de control)
Sistema que tiene intercambio de masa y energía con el medio (Como el que se muestra en la imagen de la turbina).

- ### Sistema aislado 
Sistema que no tiene intercambio de masa o energía con el medio (poco practico, salvo a análisis particulares para ciertos compuestos).

- ## Medio
Todo aquello que no hacer parte del sistema

- ## Frontera
Interfaz que separa el sistema del medio

- ## Equilibrio (Equilibrio Térmico)
Estado termodinamico en el cual ninguna de sus propiedades cambian significativamente a medida que transcurre el tiempo.
Para que se cumpla el equilibrio Térmico debe de cumplirse equilibrio :
- ### ...Mecánico :
Sumatoria de fuerzas internas y con el medio igual a cero (contribuida por presión).
- ### ...Químico
No se presentan reacciones químicas, producción/desaparición de nuevas especies
- ### ...Térmico
No se presentan cambios de temperatura y energía interna.
- ### ...Fase
No se presentan cambios de fase, incluye tambien el equilibrio dinámico en este aspecto, en el que si una molecula escapa a la fase gaseosa, otra de la fase gaseosa ingresa a la fase líquida

- ## Estado
Condición de un sistema en el que tiene establecidas ciertas propiedades

- ## Propiedades:
Son las características de un sistema (Temperatura, presión, densidad, entalpía).
Tipos de propiedades:
- ### Intensivas
Son aquellas que no dependen de la cantidad de masa (como la Temperatura o presión), es decir, de un macrocuerpo, si toman una parte de ese macrocuerpo, la temperatura sera siendo igual de todo el cuerpo y no queda subdividido
- ### Extensivas
Son aquellas que si dependen de la cantidad de masa (como el volumen, energía interna), pueden convertirse en Intensivas si se divide por la masa, para que quede por unidad de masa y se considera a partir de entonces como propiedad específica. EJ. volumen es el espacio que ocupa un cuerpo, al dividir entre la masa, ya tenemos un volumne específico que queda en unidades de m³/kg 

- ## Proceso
Describe como un sistema va de un estado a otro



existen los siguientes tipos de proceso:
- ### ...Isotérmico :
Procesos a temperatura constante (ΔT = 0)



- ### ...Isobárico :
Procesos a presión constante (ΔP = 0)
- ### ...Isocórico :
Procesos a volumen constante (ΔV = 0)
- ### ...Isoentálpico :
Procesos a entalpía constante (ΔH = 0)
- ### ...Isoentrópico :
Procesos a entropía constante (ΔS = 0)

- ## Ciclo termodinámico
Conjunto de procesos en el que el punto de inicio es el punto final y se reinicia el proceso cíclico




# Temperatura:
Hace referencia al estado cinetico de un cuerpo a nivel molecular, es decir, en que condiciones esta la dinamica de interacciones moleculares a nivel interno de un cuerpo; tambien es una referencia de como esta la energía interna de ese sistema. A temperaturas altas, se presenta mayor colisión entre párticulas a nivel interno, si es mas baja, menos movimientos se presentan a nivel molecular.



## Ley cero de la térmodinamica : 
"Todos aquellos cuerpos que esten en contacto entre si, en algún momento, alcanzaran ambos la misma temperatura"
"Todos aquellos cuerpos que esten en contacto con el ambiente, en algún momento, la temperatura ambiente"

# Densidad(ρ) / volumen específico(v) :
Ya conocemos lo que es masa (de que se compone la materia a nivel molecular y atómico), y el volumen (cuanto espacio ocupa un cuerpo), al realizar la divisón entre estas dos propiedades tenemos densidad ([=] kg/m³) que indica, en unidad de volumen, cuanta masa puede ocupar



El volumen específico es el inverso de la densidad ( v = 1/ρ), que específica, una unidad de masa de un sistema, cuanto espacio ocupa
