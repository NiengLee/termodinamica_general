# Trabajo

$$
W = \int_{x_i}^{x_f} F\,\mathrm{d}x
$$


diferencias en diferenciales:
- d : cambio diferencial de una propiedad a nivel de dos estados distintos, EJ. dT es un cambio infinitesimal (impercibible) de Temperatura es decir, 20°C sube un diferencial a 20.000...0001°C
- $\delta$ : cambio diferencial de una propiedad que depende de el proceso que ejecuta (no aplica a T, porque T es una propiedad de estado, no un proceo), aplica para transferencia de calor, o trabajo
- $\Delta$ : Cambio discreto de una propiedad entre dos estados distintos, EJ ΔT es un cambio discreto (considerable) de Temperatura es decir, 20°C sube un delta a 21°C

# Calor
Transferencia de energía a nivel molecular por colisión de particulas. Siempre fluye de la fuente de alta temperatura a la baja temperatura



## Mecánismos de transferencia de calor : 



### Conducción :
$$
Q_{\text{conduccion}} = -kA\frac{\mathrm{d}T}{\mathrm{d}x} = -kA\frac{\Delta T}{L}
$$
El flujo de calor se da atraves de un cuerpo solido. (Ley de fourier).
- L : espesor del solido entre fuente alta de temperatura y la fuente baja de temperatura.
- A : area de superficie que esta en contacto con la fuente alta de temperatura.
- k : coeficiente de conducción.

### Convección : 
$$
Q_{\text{conveccion}} = hA(T_s - T_{\text{amb}})
$$
El flujo de calor se da atraves de un fluido liquido o gaseoso. (Ley de enfriamiento de Newton).
- Ts: Temperatura de la superficie.
- Tamb : Temperatura del fluido (ambiente). 
- h : coeficiente de convección.

### Radiación : 
$$
Q_{\text{radiacion}} = \epsilon \sigma A (T_{s}^{4} - T_{rec}^{4})
$$
El flujo de calor se da atraves del vacio. (Ley Stefan-Boltzmann).
- ε : emisividad
- σ : constante de stefan-boltzmann
- Trec : Temperatura del recinto

# Energía :
Es una propiedad estado de un sistema respecto a un punto de referencia. Propiedad que permite que el sistema pueda realizar cambios sobre si mismo o en su entorno, se puede gastar o acumular (forma, posición, rotación, vibración).

## Formas de energía:
### Energía cinética (Ec)
$$
E_{c} = \frac{m\vec{v}^2}{2}
$$
Relacionado con la velocidad de movimiento del sistema

### Energía potencial (Ep)
$$
E_{c} = mgh
$$
Relacionado con la diferencia de posición entre el sistema y un macrocuerpo

### Energía mecánica (Em)
$$
E_{mecanica} = E_{c} + E_{p}
$$
Suma de energía potencial y energía cinética

### Energía interna (U)
$$
U = m\,f(T)
$$

Relacionado con la energía generada por interacciones moleculares



# Primera ley de la termodinámica : 
"la masa y la energía en el universo son constante, no se crea ni se destruye"

## Para sistemas cerrados : 

$$
\mathrm{d}E = \delta W + \delta Q
$$

$$
\frac{\mathrm{d}E}{\mathrm{d}t} = \dot{W} + \dot{Q}
$$

$$
\Delta E = W_{\text{neto}} + Q
$$

$$
\Delta U + \Delta E_c + \Delta E_p = W + Q
$$


## Terminología :

- ### Régimen estacionario :
No hay cambio de propiedades internas del sistema ($\Delta E, \mathrm{d}E = 0$)

- ### Proceso adiabático :
No hay transferencia de calor entre el medio y el sistema ($Q = 0$)

- ### Proceso cuasiestático :
Cambio de un sistema entre un estado y otro, por medio de cambios diferenciales en sus propiedades

- ### Proceso reversible :
Proceso en el cual la ruta de un proceso entre A y B es el mismo de B a A 

- ### Proceso irreversible :
Procesos no reversibles
