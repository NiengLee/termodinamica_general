<h1 style="text-align:center;">Capítulo 10 - Relaciones termodinámicas</h1>

<p style="text-align:center;">Es el fundamento teórico para determinar propiedades como u, h, s, a partir de propiedades que pueden ser directamente medibles como P, T, v. La herramienta que se usa para relacionar todas las propiedes entre se encuentran en el cálculo</p>

# Fundamentos de derivadas parciales

Sean P, T y $v$ presión, temperatura, y volumen específica. Podemos expresar una propiedad en función de las otras dos, como:

$$
P = f(T.v) \\[10pt]
T = f(P.v) \\[10pt]
v = f(P.T) \\[10pt]
$$

Trabajaremos con P (pero funciona de manera análoga con T o con v), el diferencial total es:

$$
\begin{aligned}
dP &= \left(\frac{\partial P}{\partial T}\right)_vdT + \left(\frac{\partial P}{\partial v}\right)_T dv \\[10pt]
dP &= MdT + N dv \\[10pt]
\end{aligned}
$$

tenemos entonces :

$$
\begin{aligned}
M &= \left(\frac{\partial P}{\partial T}\right)_v \\[10pt]
N &= \left(\frac{\partial P}{\partial v}\right)_T \\[10pt]
\end{aligned}
$$

ahora realizamos la siguiente operación diferencial :
$$
\begin{aligned}
\frac{\partial M}{\partial v} &= \frac{\partial^2 P}{\partial T \partial v} \\[10pt]
\frac{\partial N}{\partial T} &= \frac{\partial^2 P}{\partial v \partial T}\\[10pt]
\end{aligned}
$$

### De lo anterior se concluye la siguiente expresión (también conocida como la **Prueba de exactitud**) :
$$
\begin{aligned}
\left(\frac{\partial M}{\partial v}\right)_T &= \left(\frac{\partial N}{\partial T}\right)_v \\[10pt]
\end{aligned}
$$

### Relación de reciprocidad : 
$$
\begin{aligned}
\left(\frac{\partial T}{\partial P}\right)_v &= \frac{1}{\left(\frac{\partial P}{\partial T}\right)_v \\[10pt]}
\end{aligned}
$$

### Relación de cíclica : 
$$
\begin{aligned}
\left(\frac{\partial P}{\partial T}\right)_v \left(\frac{\partial T}{\partial v}\right)_P \left(\frac{\partial v}{\partial P}\right)_T = -1 \\[10pt] 
\end{aligned}
$$

Las ecuaciones anteriores son deducidas por operar y sustituir con $T = f(P.v)$ y $v = f(P.T)$ sus expresiones derivadas

# Relaciones fundamentales entre propiedades :

Por primera ley:

$$
\mathrm{d}u = \delta q + \delta w
$$

y tambien sabemos que :

$$
\delta w = -P\mathrm{d}v \\[10pt]
\delta q = T\mathrm{d}s \\[10pt]
$$

Tenemos una expresión para energía interna:

$$
\checkmark \mathrm{d}u = T\mathrm{d}s - P\mathrm{d}v
$$

Ahora con entapía $h = u + Pv$ y derivando : $\mathrm{d}h = \mathrm{d}u + v\mathrm{d}P + P\mathrm{d}v$ reemplazamos en la ecuación el termino du :

$$
\mathrm{d}h - v\mathrm{d}P - \cancel{P\mathrm{d}v} = T\mathrm{d}s - \cancel{P\mathrm{d}v} \\[10pt]
\checkmark \mathrm{d}h = T\mathrm{d}s + v\mathrm{d}P \\[10pt]
$$

ahora tenemos **función de Helmholtz(a)** : 

$$
a = u - Ts
$$

derivando la función anterior y reemplazando con al ecuacion de du :

$$
\checkmark \mathrm{d}a = -P\mathrm{d}v - s\mathrm{d}T
$$

y finalmente, definimos la  **función de Gibbs(g)** :

$$
g = h - Ts
$$

derivando la función anterior y reemplazando con al ecuacion de dh :

$$
\checkmark \mathrm{d}g = v\mathrm{d}P - s\mathrm{d}T
$$

# Resumen de las ecuaciones de Gibbs :

$$
\begin{aligned}
\checkmark \mathrm{d}u &= T\mathrm{d}s - P\mathrm{d}v \\[10pt]
\checkmark \mathrm{d}h &= T\mathrm{d}s + v\mathrm{d}P \\[10pt]
\checkmark \mathrm{d}a &= -P\mathrm{d}v - s\mathrm{d}T \\[10pt]
\checkmark \mathrm{d}g &= v\mathrm{d}P - s\mathrm{d}T \\[10pt]
\end{aligned}
$$

# Relaciones de Maxwell :

con las 4 expresiones anteriores y lo que se planteo al inicio tomando el concepto de la **Prueba de exactitud** :

$$
\begin{aligned}
dP &= MdT + N dv \\[10pt]
\left(\frac{\partial M}{\partial v}\right)_T &= \left(\frac{\partial N}{\partial T}\right)_v \\[10pt]
\end{aligned}
$$

Tenemos las expresiones de **las relaciones de Maxwell** :

$$
\begin{aligned}
\checkmark \left(\frac{\partial T}{\partial v}\right)_s &= - \left(\frac{\partial P}{\partial s}\right)_v \\[10pt]
\checkmark \left(\frac{\partial T}{\partial P}\right)_s &= \left(\frac{\partial v}{\partial s}\right)_P \\[10pt]
\checkmark \left(\frac{\partial P}{\partial T}\right)_v &= \left(\frac{\partial s}{\partial v}\right)_T \\[10pt]
\checkmark \left(\frac{\partial v}{\partial T}\right)_P &= - \left(\frac{\partial s}{\partial P}\right)_T \\[10pt]
\end{aligned}
$$

# Relaciones generalizadas para Entropía:

Recordemos que:

$$
\begin{aligned}
\checkmark \mathrm{d}u &= T\mathrm{d}s - P\mathrm{d}v \\[10pt]
\checkmark \mathrm{d}h &= T\mathrm{d}s + v\mathrm{d}P \\[10pt]
\end{aligned}
$$

derivando parcialmente respecto a temperatura volumen constate para la primera expresión y a presión constante para la segunda expresión:

$$
\begin{aligned}
\left(\frac{\partial u}{\partial T}\right)_v &= T\left(\frac{\partial s}{\partial T}\right)_v \\[10pt]
\left(\frac{\partial h}{\partial T}\right)_P &= T\left(\frac{\partial s}{\partial T}\right)_P \\[10pt]
\end{aligned}
$$

y recordamos las siguientes expresiones para capacidades caloríficas:

$$
\begin{aligned}
\left(\frac{\partial u}{\partial T}\right)_v &= C_v \\[10pt]
\left(\frac{\partial h}{\partial T}\right)_P &= C_P \\[10pt]
\end{aligned}
$$

port tanto:

$$
\begin{aligned}
T\left(\frac{\partial s}{\partial T}\right)_v &= C_v \rightarrow \left(\frac{\partial s}{\partial T}\right)_v &= \frac{C_v}{T} \\[10pt]
T\left(\frac{\partial s}{\partial T}\right)_P &= C_P \rightarrow \left(\frac{\partial s}{\partial T}\right)_P &= \frac{C_P}{T} \\[10pt]
\end{aligned}
$$

tomamos entropía como función de temperatura ($T$) y volumen ($v$) :

$$
s = f(T,v) \\[10pt]
ds = \left(\frac{\partial s}{\partial T}\right)_v \mathrm{d}T + \left(\frac{\partial s}{\partial v}\right)_T \mathrm{d}v \\[10pt]
$$

de lo anterior, reemplazamos el componente de capacidad calorifica a volumen constante, y una de las relaciones de Maxwell, para volumen constante:

$$
\checkmark ds = \frac{C_v \mathrm{d}T}{T}  + \left(\frac{\partial P}{\partial T}\right)_v \mathrm{d}v \\[10pt]
$$

ahora si tomamos entropía como función de temperatura ($T$) y presión ($P$), de manera análoga :

$$
s = f(P,T) \\[10pt]
\checkmark ds = \frac{C_p \mathrm{d}T}{T}  - \left(\frac{\partial v}{\partial T}\right)_P \mathrm{d}P \\[10pt]
$$

## Resumen para las expresiones generalizadas para entropía : 

$$
\checkmark ds = \frac{C_v \mathrm{d}T}{T}  + \left(\frac{\partial P}{\partial T}\right)_v \mathrm{d}v \\[10pt]
\checkmark ds = \frac{C_p \mathrm{d}T}{T}  - \left(\frac{\partial v}{\partial T}\right)_P \mathrm{d}P \\[10pt]
$$

# Relaciones generalizadas para entalpía y energía interna :

sabemos que :

$$
\begin{aligned}
\mathrm{d}u &= T\mathrm{d}s - P\mathrm{d}v \\[10pt]
\mathrm{d}h &= T\mathrm{d}s + v\mathrm{d}P \\[10pt]
\end{aligned}
$$

y también que :

$$
\begin{aligned}
ds &= \frac{C_v \mathrm{d}T}{T}  + \left(\frac{\partial P}{\partial T}\right)_v \mathrm{d}v \\[10pt]
ds &= \frac{C_p \mathrm{d}T}{T}  + \left(\frac{\partial v}{\partial T}\right)_P \mathrm{d}P \\[10pt]
\end{aligned}
$$

sustituimos ds en las primeras expresiones y reorganizamos algebraicamente:

$$
\begin{aligned}
\checkmark \mathrm{d}u &= C_vdT + \left[T\left(\frac{\partial P}{\partial T}\right)_v - P \right]\mathrm{d}v \\[10pt]
\checkmark \mathrm{d}h &= C_PdT + \left[v - T\left(\frac{\partial v}{\partial T}\right)_P \right]\mathrm{d}P \\[10pt]
\end{aligned}
$$


# Relaciones generalizadas para $C_P$ y $C_v$

Partimos de lo siguiente:

$$
ds = \frac{C_v \mathrm{d}T}{T}  + \left(\frac{\partial P}{\partial T}\right)_v \mathrm{d}v \\[10pt]
ds = \frac{C_p \mathrm{d}T}{T}  - \left(\frac{\partial v}{\partial T}\right)_P \mathrm{d}P \\[10pt]
$$

tomando el concepto de la **Prueba de exactitud** :

$$
\begin{aligned}
dP &= MdT + N dv \\[10pt]
\left(\frac{\partial M}{\partial v}\right)_T &= \left(\frac{\partial N}{\partial T}\right)_v \\[10pt]
\end{aligned}
$$

Tenemos las siguientes expresiones :

$$
\begin{aligned}
\left(\frac{\partial C_v}{T \partial v}\right)_T &= - \left(\frac{\partial^2 P}{T\partial T^2}\right)_v \rightarrow \left(\frac{\partial C_v}{\partial v}\right)_T = - T\left(\frac{\partial^2 P}{\partial T^2}\right)_v\\[10pt]
\left(\frac{\partial C_P}{T \partial P}\right)_T &= - \left(\frac{\partial^2 v}{\partial T^2}\right)_P \rightarrow \left(\frac{\partial C_P}{\partial P}\right)_T = - T\left(\frac{\partial^2 v}{\partial T^2}\right)_P\\[10pt]
\end{aligned}
$$

Ahora, igualando las dos expresiones iniciales $\mathrm{d}s$, asumineod que no hay cambio de volumen, es decir, procesos isocóricos y tenemos:

$$
\checkmark C_P - C_v = T \left(\frac{\partial v}{\partial T}\right)_P \left(\frac{\partial P}{\partial T}\right)_v
$$

y por relación cíclica:

\begin{aligned}
\left(\frac{\partial P}{\partial T}\right)_v \left(\frac{\partial T}{\partial v}\right)_P \left(\frac{\partial v}{\partial P}\right)_T = -1 \\[10pt] 
\left(\frac{\partial P}{\partial T}\right)_v = - \left(\frac{\partial v}{\partial T}\right)_P \left(\frac{\partial P}{\partial v}\right)_T \\[10pt] 
\end{aligned}

finalmente obtenemos:

$$
\checkmark C_P - C_v = T \left(\frac{\partial v}{\partial T}\right)^2_P \left(\frac{\partial P}{\partial v}\right)_T
$$


# Variación de propiedades en sólidos y líquidos :

## Coeficiente de expación volumétrico $\beta$

$$
\beta = \frac{1}{v} \left(\frac{\partial v}{\partial T}\right)_P
$$

## Coeficiente de compresibilidad isotermo $K_T$

$$
K_T = -\frac{1}{v} \left(\frac{\partial v}{\partial P}\right)_T
$$

tenemos entonces para $C_P - C_v$

$$
C_P - C_v = \frac{vT\beta^2}{K_T}
$$

Ahora tomando un proceso isotermico para sistemas en estado sólido o líquido:

$$
\begin{aligned}
ds &= \cancel{\frac{C_p \mathrm{d}T}{T}}  - \left(\frac{\partial v}{\partial T}\right)_P \mathrm{d}P \\[10pt]
ds &= - v\beta \mathrm{d}P \\[10pt]
\end{aligned}
$$

Reorganizando el termino de $K_T$

$$
-v K_T dP = \partial v_{T=const}
$$

usamos lo anterior para determinar el trabajo de compresión de un gas a temperatura constante (en liquidos y solidos, el volumen específico permanece constante en sólidos y líquidos):

$$
\begin{aligned}
w_t = - \int_1^2 Pdv =  \int_1^2 v K_T P dP = \frac{v K_T}{2}(P_2^2 - P_1^2)
\end{aligned}
$$

# Comportamientos en estados de saturación ; Presión de vapor -  Ecuación de Clapeyron

Estos análisis se relaizan para sistemas en los que se presentan cambios de fase, es decir, en estados de saturación.

tomamos entropía en función de temperatura y volumen, y realizamos operaciones diferenciales:

$$
ds = \left(\frac{\partial s}{\partial T}\right)_v \mathrm{d}T + \left(\frac{\partial s}{\partial v}\right)_T \mathrm{d}v \\[10pt]
$$

en cambios de fase, la temperatura es constante y teniendo presente que $\left(\frac{\partial P}{\partial T}\right)_v = \left(\frac{\partial s}{\partial v}\right)_T$

$$
ds = \cancel{\left(\frac{\partial s}{\partial T}\right)_v \mathrm{d}T} + \left(\frac{\partial P}{\partial T}\right)_v \mathrm{d}v \\[10pt]
\int_1^2 ds = \int_1^2 \left(\frac{\partial P}{\partial T}\right)_v \mathrm{d}v \\[10pt]
$$

en cambio de fase, $\left(\frac{\partial P}{\partial T}\right)_v$ no varía respecto al volumen específico, por tanto:

$$
s_2 - s_1 = \left(\frac{\partial P}{\partial T}\right)_{sat} (v_2-v_1) \\[10pt]
\left(\frac{\partial P}{\partial T}\right)_{sat} = \frac{s_2 - s_1}{v_2-v_1} \\[10pt]
$$

ahora con $dh = T ds + v dP$, dP es nulo porque en cambio de fase la presión no cambia y T es constante, entonces

$$
\left(\frac{\partial P}{\partial T}\right)_{sat} = \frac{h_2 - h_1}{T(v_2-v_1)} \\[10pt]
$$

## Ecuación de Clapeyron

La ecuación de Clapeyron aplica en la ecuación anterior para condiciones f y g (propiedades en líquido saturado y propiedades en vapor saturado), puede aplicarse tambien para sólido vapor o sólido - líquido

$$
\left(\frac{\partial P}{\partial T}\right)_{sat} = \frac{h_{fg}}{T(v_{fg})} \\[10pt]
$$

## Ecuación de Clausius Clapeyron

$$
\frac{\mathrm{d} P}{P} = \frac{h_{fg} \mathrm{d}T}{RT^2} \\[10pt]
$$

# Ecuación de Antoine
Para variaciones ya considerables de presión y temperatura, se toma la ecuación de clapeyron y se integra
$$
\int \frac{\mathrm{d} P}{P} = \int \frac{h_{fg} \mathrm{d}T}{RT^2} \\[10pt]
$$

- \frac{h_{fg}}{R} es constante, y sera igual a 

$$
\ln{P} = - \frac{h_{fg}}{RT} + C \\[10pt]
$$

- Sea A la constante de intregación
- Sea $\frac{h_{fg}}{R}$ igual a B

$$
\ln{P} = A - \frac{B}{T} \\[10pt]
$$

agregando una constante C como sumando a la temperatura, tendremos la ecuación de Antoine:

$$
\checkmark \ln{P} = A - \frac{B}{T + C} \\[10pt]
$$

Es importante tener presente que muchas veces el logaritmo natural se reemplaza por logaritmo en base 10, todo depende de la fuente en la que se toman las constantes de la expresión de Antoine como referencia:

$$
\checkmark \log_{10}P_{sat} = A - \frac{B}{T_{sat} + C} \\[10pt]
$$

La expresión anterior solo funciona para calcular la presión de vapor para una temperatura específica

# Coeficiente de Joule-Thomson

Aplica para procesos de estrangulamiento de flujo, donde generalmente los procesos son adiabáticos y por tanto isoentálpicos, debe de tenerse presente que aplica principalmente a sistemas como:

- Válvulas de expansión.
- Sistemas de refrigeración.
- Procesos de licuefacción de gases.
- Reguladores de presión.
