La siguiente ecuación de estado aplica para gases en condiciones ideales:

- Presiones bajas
- Temperaturas altas
- Gases mono/di-atomicos
- Siempres es mejor consultar por tablas de propiedades antes de recurrir a esta expresión

$$
Pv = \eta R_{\text{gas}}T
$$

- Presión
- Volumen
- Número de moles
- Constante de gases
- Temperatura

$$
\overline{PM} = \frac{m}{\eta}
$$

- Masa
- Número de moles

$$
\rho_g = \frac{P.\overline{PM}}{R_{\text{gas}}T}
$$

Valores para $R_{\text{gas}}$ :

- $8.3145 \: [=] \: \frac{Pa.m³}{mol.K} \: [=] \: \frac{J}{mol.K} \: [=] \: \frac{kJ}{kmol.K}$
- $0.0083145 \: [=] \: \frac{kJ}{mol.K}$


# Capacidades caloríficas :
Indica la cantidad de energía que consume o desprende al cambiar un grado de temperatura para solo una unidad de masa

- $C_v$ = capacidad calorífica a volumen constante
- $C_p$ = capacidad calorífica a presión constante

las unidades estan normalmente en $\frac{kJ}{kg.K}$ pero tambien pueden estar en $\frac{kJ}{kmol.K}$ o $\frac{kJ}{kmol.°C}$ y sus combinaciones

## Definición formal de entalpía :
Como se puede observar a continuación, tiene en cuenta la energía a nivel molecular de un sistema a nivel interno, pero, tambien se debe tener en cuenta el trabajo mecánico que se ejerce sobre ese sistema (en nuestro caso fluido), que contribuye al movimiento a nivel macro de este sistema :

$$
h = u + Pv
$$

- Esto es importante tenerlo en cuenta ya que, en la energía interna se considera para sistemas cerrados y evaluar las interacciones moleculares y como afectan su energía total. 
- En sistemas abiertos, ademas de tener interacciones moleculares que afectan su valor energético, tambien se debe tener presente el trabajo mecanico a la que se somete el sistema para moverse de un punto a otro (parecido a gas en un piston)

## Expresiones para capacidades caloríficas : 

- Sea $u = f(T,v)$ y $h = f(T,P)$
- $u$ no varia considerablemente al cambiar $v$ ($\left(\frac{\partial u}{\partial v}\right)_{T=\text{cte}} = 0$)
- Para $h$ el proceso de cambio de un estado a otro se realiza a presión constante ($\mathrm{d}P = 0$)

$$
\mathrm{d}u = C_v(T)\mathrm{d}T \: = \: \int_{T_i}^{T_f} C_v(T)\,\mathrm{d}T
$$
$$
\mathrm{d}h = C_p(T)\mathrm{d}T \: = \: \int_{T_i}^{T_f} C_p(T)\,\mathrm{d}T
$$

$C_v(T)$ y $C_p(T)$ pueden encontrarse en referencias bibliográficas como expresiones polinómicas en función de la temperatura.



## Relacion entre $C_v(T)$ y $C_p(T)$
- Solo funciona para gases
- $v$ es el volumen específico molar, esto para anular el termino $\eta$
- Expresión util para el caso en el que solo se desea conocer el cambio de energía interna de un sistema, pero solo se dispone de $C_p$

$$
h = u + Pv \rightarrow \mathrm{d}h = \mathrm{d}u + \mathrm{d}(Pv) \: ; \: Pv = RT
$$

$$
\mathrm{d}h = \mathrm{d}u +  R\mathrm{d}T
$$

$$
C_p(T)\mathrm{d}T = C_v(T)\mathrm{d}T + R\mathrm{d}T
$$

$$
C_p(T) = C_v(T) + R
$$

Muchas veces se expresa $\gamma = \frac{C_p}{C_v}$ por tanto:

$$
\gamma = 1 + \frac{R}{C_v}
$$

# Estado de líquido subenfriado :

Todo lo visto hasta ahora aplica para estado de vapor sobrecalentado o gases y da resultados aproximados, pero para condiciones de liquido subenfriado, se deben tener algunos detalles en cuenta

- La ecuación de gas ideal ya no aplica (NO CALCULAR PRESIÓN DE UN LIQUIDO CON ECUACION DE GAS IDEAL)
- $\rho$ (tambien el $v$) no cambia significativamente con el cambio de presión o temperatura en estado líquido (son valores constantes)
- la capacidad calorífica para liquidos es estandar, no se diferencia si es a presió o volumen constante
- las propiedades de líquidos debajo del punto de saturación, se pueden aproximar a estado de saturación

$$
h = u + Pv \rightarrow \mathrm{d}h = \mathrm{d}u + \cancel{\mathrm{d}(Pv)}
$$

$$
\mathrm{d}h = \mathrm{d}u
$$

$$
C_p(T)\mathrm{d}T = C_v(T)\mathrm{d}T
$$

$$
C_p(T) = C_v(T) 
$$

$$
\Delta u = \int_{T_i}^{T_f} C(T)dT\\[10pt]
\Delta h = \int_{T_i}^{T_f} C(T)dT + v\Delta P
$$
